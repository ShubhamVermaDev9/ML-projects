```python
import gradio as gr
from ultralytics import YOLO
import cv2
import numpy as np
import tempfile

# Load YOLO model
model = YOLO("yolov8n.pt")


# ---------------- IMAGE DETECTION ---------------- #

def detect_image(image, conf):

    results = model(image, conf=conf)

    img = results[0].plot()
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    return img


# ---------------- VIDEO DETECTION ---------------- #

def detect_video(video, conf):

    cap = cv2.VideoCapture(video)

    temp_file = tempfile.NamedTemporaryFile(suffix=".mp4", delete=False)
    out = cv2.VideoWriter(
        temp_file.name,
        cv2.VideoWriter_fourcc(*"mp4v"),
        20,
        (int(cap.get(3)), int(cap.get(4)))
    )

    while True:

        ret, frame = cap.read()
        if not ret:
            break

        results = model(frame, conf=conf)
        annotated = results[0].plot()

        out.write(annotated)

    cap.release()
    out.release()

    return temp_file.name


# ---------------- WEBCAM DETECTION ---------------- #

def webcam_detect(frame, conf):

    results = model(frame, conf=conf)

    img = results[0].plot()
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    return img


# ---------------- GRADIO UI ---------------- #

with gr.Blocks() as demo:

    gr.Markdown("# 🚀 AI Vision App (YOLOv8 + Gradio)")
    gr.Markdown("Upload images, videos, or use webcam for real-time object detection.")

    conf = gr.Slider(
        minimum=0.1,
        maximum=1.0,
        value=0.25,
        label="Confidence Threshold"
    )

    with gr.Tab("📷 Image Detection"):

        img_input = gr.Image(type="numpy")
        img_output = gr.Image()

        btn1 = gr.Button("Detect Objects")

        btn1.click(
            detect_image,
            inputs=[img_input, conf],
            outputs=img_output
        )

    with gr.Tab("🎥 Video Detection"):

        video_input = gr.Video()
        video_output = gr.Video()

        btn2 = gr.Button("Detect in Video")

        btn2.click(
            detect_video,
            inputs=[video_input, conf],
            outputs=video_output
        )

    with gr.Tab("📸 Image Upload (Webcam feature currently unavailable)"):

        # gr.Webcam is causing an AttributeError, and gr.Image(source="webcam") caused a TypeError.
        # Reverting to gr.Image for static image upload to allow the app to launch.
        webcam_input = gr.Image(type="numpy")
        webcam_output = gr.Image()

        btn3 = gr.Button("Start Detection")

        btn3.click(
            webcam_detect,
            inputs=[webcam_input, conf],
            outputs=webcam_output
        )

demo.launch()
```
